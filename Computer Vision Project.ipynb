```python
#Project Code

import cv2
import numpy as np
import mediapipe as mp
import time

#media pipe
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.7)
mp_drawing = mp.solutions.drawing_utils

#canvas
canvas = np.zeros((480, 640, 3), dtype=np.uint8)

#  capturing 
cap = cv2.VideoCapture(0)

# flag to track if the image should be saved
save_image = False

def finger_status(hand_landmarks):
    """Determine if specific fingers are up."""
    thumb_is_up = hand_landmarks.landmark[4].y < hand_landmarks.landmark[3].y
    index_finger_is_up = hand_landmarks.landmark[8].y < hand_landmarks.landmark[6].y
    small_finger_is_up = hand_landmarks.landmark[20].y < hand_landmarks.landmark[18].y
    two_fingers_up = (hand_landmarks.landmark[8].y < hand_landmarks.landmark[6].y and 
                      hand_landmarks.landmark[12].y < hand_landmarks.landmark[10].y)
    return index_finger_is_up, two_fingers_up, thumb_is_up,small_finger_is_up

def mouse_click(event, x, y, flags, param):
    """Capture mouse click event to save image."""
    global save_image
    if event == cv2.EVENT_LBUTTONDOWN:
        save_image = True
        print("Mouse clicked, saving image...")

# mouse callback function
cv2.namedWindow("Virtual Calculator")
cv2.setMouseCallback("Virtual Calculator", mouse_click)

while True:
    success, frame = cap.read()
    if not success:
        break
    
    frame = cv2.flip(frame, 1)  # fliping the image horizontally for a mirror view
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = hands.process(img_rgb)

    # checking hand landmarks
    if result.multi_hand_landmarks:
        for hand_landmarks in result.multi_hand_landmarks:
            mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
            
            #  finger statuses
            index_finger_is_up, two_fingers_up, thumb_is_up, small_finger_is_up = finger_status(hand_landmarks)
            
            #  actions based on gestures- perform
            if index_finger_is_up and not two_fingers_up:
                # drawing on canvas when index finger is up
                x, y = int(hand_landmarks.landmark[8].x * 640), int(hand_landmarks.landmark[8].y * 480)
                cv2.circle(canvas, (x, y), 5, (255, 255, 255), -1)
            elif two_fingers_up:
                # navigation mode (placeholder)
                print("Navigate canvas mode")
            elif thumb_is_up:
                # reset/erase 
                canvas.fill(0)
                print("Canvas reset")
            elif small_finger_is_up:
                pass

    # combining canvas and frame
    combined = cv2.addWeighted(frame, 0.5, canvas, 0.5, 0)
    cv2.imshow("Virtual Calculator", combined)

    # saving the image when mouse click
    if save_image:
        timestamp = int(time.time())
        filename = f"equation_{timestamp}.png"
        cv2.imwrite(filename, canvas)  # saving the canvas as an image
        print(f"Image saved as {filename}")
        save_image = False  # resetting the flag after saving the image

    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
